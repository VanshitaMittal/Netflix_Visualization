```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
data_set = pd.read_csv('Netflix Shows.csv', encoding = "ISO-8859-1")
```

```python
data_set = data_set.drop_duplicates()
```

```python
data_set[data_set.duplicated('title',keep = False)]
```

```python
latest_best = data_set[(data_set['user rating score']>90) & (data_set['release year']>2016)]['title']
best = latest_best.values
for i in range(0, len(best)):
    print(best[i])
```

```python
rat = data_set['rating'].value_counts()
sns.barplot(rat.index, rat.values, alpha=0.8)
plt.xticks(rotation='vertical')
plt.xlabel('Type of rating', fontsize=14)
plt.ylabel('Number os shows', fontsize=14)
plt.title("Rating of Netflix programs", fontsize=16)
plt.show()
```

```python
yr = data_set.groupby('release year').size()
plt.figure(figsize = (10,6))
plt.plot(yr.index, yr.values, color = 'red')
plt.xlabel('Years')
plt.ylabel('No. of Movies')
plt.title('Number of shows released with respect to each year')
plt.show()
```
